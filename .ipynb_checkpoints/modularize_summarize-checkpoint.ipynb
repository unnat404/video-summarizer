# function to summarize video

```python
# score (online) = 36.92535
import speech_recognition as sr 
from moviepy.editor import *
import moviepy.editor as mp
# gensim==3.8.3
import gensim
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
import json


# NOTE : AUDIO files should be ".wav" and of no other format will work

#function to convert video to text
def video_to_text(videoclip):
    # C:\Users\UNNAT\summarize_hackerearth\hackerearth dataset\dataset
    # write the audio of the videoclip into the given file
    videoclip.audio.write_audiofile(r"audio_func.wav")     
    audioclip=AudioFileClip(r"audio_func.wav")
    
    duration=int(audioclip.duration)
    print(duration)
    result=""
    
    # divide video into smaller chunks to run speech to text recognizer
    # then combine the results
    for start in range(0,duration,120):
        print("\n next batch....\n")
        end=min(start+120,duration)
        sub_audio=audioclip.subclip(start,end)
        sub_audio.write_audiofile(r"temp_voiceover.wav")

        r = sr.Recognizer()
        audio = sr.AudioFile(r"temp_voiceover.wav")

        with audio as source:
            audio_file = r.record(source)
            try:
                result = result + r.recognize_google(audio_file)
                print(" Working On...")
    #             print(result)
            except:
                print('Sorry... run again....')

        if end>=duration:
            break
        
    print("\t\tExtracted text: \n\n"+result+"\n") 
    return result 
    
def summarizer(text):    
    # convert text to sentence
    result=text
    res=result.split()
    count=0
    dot_text=""
    for word in res:
        count=count+1
        dot_text = dot_text +" "+ word
        if count==13:
            # we assune the avg sentence length to be 13
            dot_text= dot_text+".\n "
            count=0
            
    print("\t\tExtracted Sentences :\n\n"+dot_text+"\n")

    # convert text to summary
    # problem: finding sentences(full stop & other punctuators)

    # extractive sumnmarization (no words are produced of its own) 
    return (summarize(dot_text,ratio=0.15))

```

# calling function with apt video location and storing in json

```python
####################################### calling function #####################################################
# NOTE : path of your video will change based on the location of your video's location 

# video1
videoclip1 = mp.VideoFileClip(r"C:\Users\UNNAT\video_summarize_hackerearth\hackerearth dataset\dataset\video1.mp4")
# videoclip = mp.VideoFileClip(r"<path_to_video>")

# 1st call function : video_to_text(videoclip) , to convert video to text 
# you can comment the next line once you have generated the text from video 
# (so as to save time and computation , while adjusting the hyperparameters for summarize function)
vid_text1=video_to_text(videoclip1)

# 2nd call function : summarizer(vid_text1) , to extract summary of text
summarized_text1=summarizer(vid_text1)

print("\n------------------\n")
print("\t\tSummarized text from function")
print(summarized_text1)
print("------------------\n")


    
# video2
videoclip2 = mp.VideoFileClip(r"C:\Users\UNNAT\video_summarize_hackerearth\hackerearth dataset\dataset\video2.mp4")
# videoclip = mp.VideoFileClip(r"<path_to_video>")

# 1st call function : video_to_text(videoclip) , to convert video to text
# you can comment the next line once you have generated the text from video 
# (so as to save time and computation , while adjusting the hyperparameters for summarize function)
vid_text2=video_to_text(videoclip2) 
# 2nd call function : summarizer(vid_text1) , to extract summary of text
summarized_text2=summarizer(vid_text2)

print("\n------------------\n ")
print("\t\tSummarized text from function")
print(summarized_text2)
print("------------------\n")

################################################# storing in json  ###################################################


final_ans={
    "Video_1":summarized_text1,
    "Video_2":summarized_text2
}
print("\n\t\t::Summary of videos in json format::\n\n")
print(final_ans)


# the json file where the output must be stored 
out_file = open("final_submission.json", "w")   

json.dump(final_ans, out_file, indent = 6)
    
out_file.close() 
# score = 36.92535
# (ratio=0.15, sentent(word_count=13))
```
